```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 80)
v = np.linspace(0, np.pi, 80)

# create the sphere surface
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z =  np.outer(np.ones(np.size(u)), np.cos(v)) + 4

# simulate heat pattern (striped)
myheatmap = (4*x*y + z**3 - 18*z**2 + 96*z)/192

ax.plot_surface(x, y, z, cstride=1, rstride=1, facecolors=cm.cool(myheatmap))

plt.show()
```

```python
def f(x, y, z):
   return x**2 + y**2 + z**2 - 8*z
	
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
z = np.linspace(-1, 9, 30)

X, Y, Z = np.meshgrid(x, y, z)
W = f(X, Y, Z)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, W, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```
