```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_range = range(-8,9)
y_range = range(-8,9)
num = 500

lines = np.empty((len(x_range)+len(y_range), 2, num))

for i in x_range: # vertical lines
    linspace_x = np.linspace(x_range[i], x_range[i], num)
    linspace_y = np.linspace(min(y_range), max(y_range), num)
    lines[i] = (linspace_x, linspace_y)
for i in y_range: # horizontal lines
    linspace_x = np.linspace(min(x_range), max(x_range), num)
    linspace_y = np.linspace(y_range[i], y_range[i], num)
    lines[i+len(x_range)] = (linspace_x, linspace_y)
    
def affine(z):
    z[:, 0] = z[:, 0] + z[:, 1] * 0.3 # transforming the x coordinates
    z[:, 1] = 0.5 * z[:, 1] - z[:, 0] * 0.8 # transforming the y coordinates
    return z

def non_affine(z):
    z[:, 0] = z[:, 0] * np.cos(z[:, 1])
    z[:, 1] = np.cos(z[:, 0]) - z[:, 1]
    return z

transformed_lines = non_affine(lines)

def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))

# bent_lines = sigmoid(transformed_lines)
bent_lines = transformed_lines


plt.figure(figsize=(8,8))
plt.axis("off")
for line in bent_lines:
    plt.plot(line[0], line[1], linewidth=1, color="k")
plt.show()
```
