```julia
using Plots
pyplot()
```

```julia
using QuadGK
```

```julia
function v(t)
    if t >= 0 && t < 5
        return -20
    elseif t >= 5 && t < 15
        return 10
    elseif t >= 15 && t < 25
        return 5
    elseif t >= 25 && t <= 30
        return -5
    else 
        return 0
    end
end
```

```julia
t = linspace(0,30,300)
plot(t,v.(t))
```

```julia
quadgk(v, 0, 30)
```

```julia
speed(t) = abs(v(t))
plot(t, speed.(t))
```

```julia
quadgk(speed, 0, 30)
```

```julia
function disp(t)
    return quadgk(v, 0, t)[1]
end
```

```julia
plot(t, disp.(t))
plot!(t, v.(t))
```

```julia
function dist(t)
    return quadgk(speed, 0, t)[1]
end
```

```julia
plot(t, dist.(t))
plot!(t, speed.(t))
```
