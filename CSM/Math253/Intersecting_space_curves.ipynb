```python
%matplotlib notebook
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(6,8))
ax = fig.gca(projection='3d')

t1 = np.linspace(-1.1, 0.1, 100)
t2 = np.linspace(2.9, 4.1, 100)

x1 = t1
y1 = 1 - t1
z1 = 3 + t1**2

x2 = 3 - t2
y2 = t2 - 2
z2 = t2

ax.plot(x1, y1, z1, label='parametric curve')
ax.plot(x2, y2, z2)
ax.legend()

plt.show()
```
