```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from numpy import *
```

```python
import mpl_toolkits.mplot3d.axes3d as axes3d
```

```python
def CentRect(cx, cy, width, height):
    lbc = cx - 0.5*width
    return patches.Rectangle((lbc,cy), width, height, edgecolor='red', alpha=0.5)
```

```python
def AreaBet(a, b, f, g, n):
    dx = (b-a)/n
    xi = a + 0.5*dx
    rects = []
    sum = 0
    for i in range(n):
        height = g(xi)-f(xi)
        rects.append(CentRect(xi, f(xi), dx, height))
        sum += height*dx
        xi += dx
        
    return sum, rects
```

```python
def plot_area_bet(a, b, f, g, n):
    sum, rects = AreaBet(a, b, f, g, n)
    
    for i in range(n):
        ax.add_patch(rects[i])
        
#     plt.text(0.25, .2, "area = ")
    print(sum)
```

```python
def f(t):
    return 0.73*t**3 - 2*t**2 + t + 0.6

def g(t):
    return 0.17*t**2 - 0.5*t + 1.1

t = linspace(0,2,100);

fig = plt.figure()
ax = fig.add_axes([0,0,2,1])

plt.plot(t, g(t))
plt.plot(t, f(t))

plot_area_bet(0, 2, f, g, 50)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,-1.5,pi,1.5])

ts = linspace(0,pi,200);

f = lambda t: sin(t)
g = lambda t: cos(t)

plt.plot(ts, [f(ti) for ti in ts])
plt.plot(ts, [g(ti) for ti in ts])

plot_area_bet(0, pi, f, g, 30)
```
