```julia
using Plots
pyplot()
```

```julia
x = range(-2*pi, stop=2*pi, length=500)
```

```julia
y = sum([(-1)^k .* x.^(2*k)./factorial(big(2*k)) for k in 0:30]);
```

```julia
y1 = sum([(-1)^k .* x.^(2*k+1)./factorial(big(2*k+1)) for k in 0:30]);
```

```julia
plot(x,y, lw=4)
```

```julia
plot!(x,y1, lw=4)
```

```julia
x = range(-3, stop=3, length=500)
y = sum([(-1)^k .* x.^(2*k+1)./((2*k+1)*factorial(big(k))) for k in 0:30]);
```

```julia
plot(x,y, lw=5)
```
