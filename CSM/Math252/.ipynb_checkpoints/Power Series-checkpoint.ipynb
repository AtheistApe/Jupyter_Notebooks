```python
import math
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def myrange(n):
    return map(lambda x: 1 + x, range(n))
```

```python
def fn(x,n):
    return (x-3)**(n+1)/(n+1)
```

```python
def f(x):
    psum = 0
    dpsum = 1
    nterms = 10
    while abs(dpsum - psum) > 0.001:
        psum = sum(fn(x,n) for n in range(nterms))
            
        dpsum = psum + sum(fn(x,nterms+n) for n in range(nterms))
        
        nterms = 2*nterms
#        if abs(dpsum-psum) > 100000:
#            psum = 0
#           break
        
    return psum
    
```

```python
x = np.arange(2,4,0.01)
```

```python
fx = list(map(f, x))
```

```python
plt.plot(x,fx)
```

```python
fx = list(map(f,x))
```
