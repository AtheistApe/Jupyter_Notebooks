```python
largest = None
smallest = None
while True:
    num = (input("Enter a number: "))
    if num == "done":
        break
    else:    
        try:
            num = int(num)
        except ValueError:
            print("Invalid input")
            continue

        if largest == None and smallest == None:
            largest = num
            smallest = num           
        else:           
            if num > largest: 
                largest = num
            if num < smallest:
                smallest = num        

print("Maximum is", largest)
print("Minimum is", smallest)
```
