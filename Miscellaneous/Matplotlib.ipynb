```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matplotlib.use(PyQt4)
%matplotlib notebook

# sample points of (t,t^2,t^3)
T = np.linspace(0, 5, 100)
X = T
Y = T**2
Z = T**3

zs = np.zeros(100)
os = np.ones(100)

# open a new figure window
F = plt.figure()
# get current axes of the figure
A = F.gca(projection='3d')
# plot the sampled points
A.plot(X, Y, Z, label='the twisted cubic',lw=3)
A.plot(X, Y, zs)
A.plot(X, 25*os, Z)
# show the label
A.legend()
# render the plot in the window
# plt.show()

XR = np.arange(0, 5, 0.01)
YR = np.arange(0,25, 0.01)
XX, YY = np.meshgrid(XR, YR)
ZZ = XX**3
S = A.plot_surface(XX, YY, ZZ, alpha=0.5)
plt.show()
```
