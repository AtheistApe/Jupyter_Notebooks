# Controlling the Mouse
https://pyautogui.readthedocs.org

```python
import pyautogui
```

```python
pyautogui.size() # Returns screen size
```

```python
width, height = pyautogui.size()
```

```python
pyautogui.position() # Returns the current cursor position
```

```python
pyautogui.moveTo(10,10) # Moves cursor to input (x,y) position
```

```python
pyautogui.moveTo(50,100, duration=2) # Moves cursor in 'duration' seconds
```

```python
pyautogui.moveTo(300,700, duration=10)
```

```python
pyautogui.moveRel(200,350, duration=2) # Moves cursor relative to current cursor position.
```

```python
cur_loc = pyautogui.position() # Find location of a clickable object.
```

```python
# Can also 'doubleClick', 'rightClick', 'leftClick' and 'middleClick'. If don't pass arguments, will click at current location.

pyautogui.click(cur_loc) # click at the object at position 'cur_loc'
```

```python
pyautogui.moveTo(cur_loc,duration=2)
```

```python
# Place cursor in a drawing program and run this program to draw a square spiral shape.

pyautogui.click()
distance = 200
while distance > 0:
    print(distance, 0)
    pyautogui.dragRel(distance, 0, duration=distance/500)
    distance -= 25
    print(0, distance)
    pyautogui.dragRel(0, distance, duration=distance/500)
    print(-distance, 0)
    pyautogui.dragRel(-distance, 0, duration=distance/500)
    distance -= 25
    print(0, -distance)
    pyautogui.dragRel(0, -distance, duration=distance/500)
```

# Controlling the Keyboard

```python
# Click at cursor location (a window that can accept type written text), then type out "Hello world!"

pyautogui.click(); pyautogui.typewrite('Hello world!')
```

```python
# Click at cursor location (a window that can accept type written text), then type out "Hello world!" where each letter takes 'interval' seconds to type.

pyautogui.click(); pyautogui.typewrite('Hello world!', interval = 0.2)
```

```python
# Can pass a list of keyboard characters to 'typewrite'.

pyautogui.click(); pyautogui.typewrite(['a', 'b', 'left', 'b'], interval = 1)
```

```python
print(pyautogui.KEYBOARD_KEYS)
```

```python
pyautogui.press('F1') # Press a single keyboard key.
```

```python
pyautogui.hotkey('command', 't') # Press a hotkey combination.
```
