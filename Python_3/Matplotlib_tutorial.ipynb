```python
%matplotlib notebook

# Import `pyplot` from `matplotlib`
import matplotlib.pyplot as plt

# Initialize the plot
fig = plt.figure(figsize=(9,3))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)

x = [np.random.rand() for i in range(1,50)]
y = [np.random.rand() for j in range(1,50)]
ax3.scatter(x,y)

# Show the plot
plt.show()
```

```python
%matplotlib notebook

# Import `pyplot` from `matplotlib`
import matplotlib.pyplot as plt
import numpy as np

# Initialize the plot
fig = plt.figure()
# fig.suptitle("Three Plots")

ax1 = fig.add_subplot(131)
ax1.set(title="First Plot",
       xlabel="x",
       ylabel="y")

ax2 = fig.add_subplot(132)
ax2.set(title="Second Plot",
       xlabel="x",
       ylabel="y")

ax3 = fig.add_subplot(133)
ax3.set(title="Third Plot",
       xlabel="x",
       ylabel="y")

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax1.axvline(0.65)

ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

# Delete `ax2`
# fig.delaxes(ax2)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from IPython.html.widgets import interactive
from IPython.display import display
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0, .25)
ax.set_ylim(-2.5, 2.5)
ax.set_title('beat frequencies')
lnA, = ax.plot([], [], color='r', label='A')
lnB, = ax.plot([], [], color='purple', label='B')
lnsum, = ax.plot([], [], color='k', label='signal')
ax.legend()
max_time = 3
rate = 8000
times = np.linspace(0,max_time,rate*max_time)

def beat_freq(f1=220.0, f2=224.0):

    A = np.sin(2*np.pi*f1*times)
    B = np.sin(2*np.pi*f2*times)
    sig = A + B

    lnA.set_data(times, A)
    lnB.set_data(times, B)
    lnsum.set_data(times, sig)

    plt.draw()

beat_freq(0, 0)
```

```python
interactive(beat_freq, f1=(200.0,300.0), f2=(200.0,300.0))
```
