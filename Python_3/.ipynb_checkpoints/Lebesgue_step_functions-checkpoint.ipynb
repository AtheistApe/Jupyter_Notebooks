```python
import numpy as np
from scipy.optimize import fmin_tnc
import math

import matplotlib.pyplot as plt
```

```python
def phi(fun, n):
    a = 0
    b = 1
    
    dx = (b-a)/2**n
    
    x = np.linspace(a, b, 2**n+1)
    y = []
    
    for j in np.arange(1,2**n+1):
        Ijl = a + (j-1)*dx
        Ijr = a + j*dx
                
        min_Ij = fmin_tnc(fun, (Ijl+Ijr)/2, approx_grad=True,
                         bounds=[(Ijl, Ijr)])
        
        ymin = fun(float(min_Ij[0]))
        y.append(ymin)
        
    y.append(fun(b))    
    plt.step(x,y, where='post')
    
#     print(dx*sum(y))
```

```python
def f(x):
    return np.sqrt(x)
```

```python
x = np.linspace(0,1,200)
```

```python
plt.figure(figsize=(12,9))
plt.plot(x, f(x))

# phi(f,0)
# phi(f,1)
phi(f,2)
# phi(f,3)
phi(f,4)
# phi(f,5)
phi(f,6)
phi(f,7)
# phi(f,10)
```
