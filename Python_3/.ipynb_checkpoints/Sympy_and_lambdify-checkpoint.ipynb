# Lambdify
This module provides convenient functions to transform sympy expressions to lambda functions which can be used to calculate numerical values very fast.

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
x,y,z,s,t = sy.symbols('x y z s t')
```

```python
fx = sy.cos(x)
f = sy.lambdify(x, fx)
```

```python
%%timeit
for t in range(10**6):
    fx.subs(x,t)
```

```python
%%timeit
for t in range(10**6):
    f(t)
```

```python
fpx = sy.diff(fx, x)
fpx
```

```python
l_fpx = sy.lambdify(x, fpx)
```

```python
l_fpx(3)
```

```python
x = np.linspace(-2*np.pi,2*np.pi,100)
y = f(x)
yp = l_fpx(x)
```

```python
plt.plot(x,y, lw=5, ls='--')
plt.plot(x,yp, lw=5, color='r')
```
