```julia
using SymPy
x = Sym("x")
```

```julia
# Must load SymPy package

# Generates the whole stack of rings.

f(x) = 2x^2 - x^3
a = 0
b = f(4/3) # Max of f at x = 4/3

function rings(n,dy)
    Δy = (b-a)/n
    ymid = [a + (i-0.5)*Δy for i in range(1,n)]
    yi = [a + i*Δy for i in range(0,n+1)]
            
    function ring(i)
        solns = sort(real(solve(f(x) - ymid[i], x)))
        rmax = solns[3]
        rmin = solns[2]
                
        str = "translate([0,0,$((i-1)*Δy)])\n{\n\tdifference()\n\t{cylinder(r=$(rmax), h=$(Δy));
            \n\tcylinder(r=$(rmin), h=$(Δy));\n}\n}"
    end
    
    for i in range(1,1,n)
        print(ring(i),"\n\n")
    end
end
```

```julia
# Must load SymPy package

# Generates each ring separately.

f(x) = 2x^2 - x^3
a = 0
b = f(4/3) # Max of f at x = 4/3


function rings2(n)
    Δy = (b-a)/n
#    dy = Δy/2 # Depth and height of binding channel/ring
#    dx = 0.05 # Width of binding channel/ring
    k = 1.5 # Scale factor for binding channel (A number slightly bigger than 1).
    
    ymid = [a + (i-0.5)*Δy for i in range(1,n)]
    yi = [a + i*Δy for i in range(0,n+1)]
    rmaxs = []
    rmins = []
    solns = []
    
    for i in range(1,n)
        solns = sort(float(real(solve(f(x) - ymid[i], x))))
        rmaxs = push!(rmaxs, solns[3])
        rmins = push!(rmins, solns[2])
    end
            
    function ri(outer, inner, height)
        str = "difference(){
                cylinder(r=$(outer), h=$(height));
                cylinder(r=$(inner), h=$(height));
                }"
        return str
    end
    
    function bind_ring(i,dx,dy)
        if i == 1
            bind_ring_radius = (rmaxs[1] + rmins[1])/2
        elseif i == n
            bind_ring_radius = (rmaxs[n] + rmins[n-1])/2
        else
            bind_ring_radius = (rmaxs[i+1] + rmins[i+1])/2
        end
        
        bro = bind_ring_radius + dx/2
        bri = bind_ring_radius - dx/2
        
        str = "translate([0,0,$(dy)]){
                $(ri(bro,bri,Δy))
                }"
        return str
    end
                
    function ring(i)
        str = "translate([0,0,$(2*(i-1)*Δy)]){
                difference(){
                    union(){
                        $(ri(rmaxs[i], rmins[i], Δy))
                        $(bind_ring(i,0.05,Δy/2))
                    }
                    
                        $(bind_ring(i-1,k*0.05,-Δy/2))
                    
                }
            }"
        return str
    end
    
    for i in range(1,n)
        print(ring(i),"\n\n")
    end
end
```
