```julia
# "partial_sum" returns a function that computes partial sums
# of the sequence function input into it.

function partial_sum(f::Function)
    n->sum([f(k) for k in 1:n])
end
```

```julia
sn = partial_sum(n -> cos(n)*sqrt(n)/(8n+5))
```

```julia
plot([sn(k) for k in 1:300])
# plot([s(n) for n in 500:1000],line = (:black, 2, 0.8))
# plot([s(n) for n in 100:300], m=(0.5, [:circle], 6))
```

```julia
sn1 = partial_sum(n->1/n^2)
```

```julia
plot([sn1(k) for k in 1:200])
```

```julia
k = 1
while abs(sn1(k+1) - sn1(k)) > 1E-7
    k += 1
end
print(sn1(k))
```
