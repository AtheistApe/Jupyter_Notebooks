```julia
using Plots
using LinearAlgebra
```

```julia
function random_data(x, f, from, to)
    unifrnd(from,to) = rand()*(to-from) + from
    return f.(x) + unifrnd(from,to)
end
```

```julia
x = 1:0.2:10
y = random_data.(x, x->2x-1,-2,2);
```

```julia
scatter_plot=scatter(x, y)
```

```julia
# Cost function
# C(m,b) = sqrt(sum((m*x.+b - y).^2))

C(m,b) = norm(m*x-y+b*ones(size(x)))
```

```julia
m = b = -10.0:0.1:10.0
z = [C(mi,bj) for mi in m for bj in b]

contour_plot=plot(m,b,z; t=:contour, levels=50)
```

```julia
# Partial derivatives of the cost function

# Cm(m,b) = sum(2.0.*x.*(m*x.+b - y))
# Cb(m,b) = sum(2.0.*(m*x.+b - y))

Cm(m,b) = sum(2.0*x'*(m*x-y+b*ones(size(x))))
Cb(m,b) = sum(2.0*(m*x-y+b*ones(size(x))))
```

```julia
step_size = 0.0001

m,b = [2, 5.0]

flag = true

while flag
    C_current = C(m,b)
    println("m: $m, b: $b, C(m,b): $(C(m,b))")
    dm,db = -step_size*[Cm(m,b), Cb(m,b)]
    m,b = m+dm, b+db
    C_new = C(m,b)
    if abs(C_current - C_new) < 1E-6
        flag = false
    end
end
```

```julia
f(x) = m*x+b

println("m = $m, b = $b")
plot(scatter_plot)
plot!(x,f)
```

```julia
function parameters(x,y)
    n = length(x)
    m = (n*dot(x,y) - sum(x)*sum(y)) / (n*dot(x,x) - (sum(x))^2)
    b = (dot(x,x)*sum(y) - dot(x,y)*sum(x)) / (n*dot(x,x) - (sum(x))^2)
    
    return (m,b)
end
```

```julia
x = 0:0.1:4
y = 0.5*x .- 2.5;
```

```julia
m,b = parameters(x,y)
```

```julia
y = random_data.(x, x->0.5*x-2.5 ,-0.5,0.5)
scatter(x,y)
plot!(x, x->m*x+b)
```

```julia
function plot_line()
    x = 1:0.2:10
    y = random_data.(x, x->x+0.5*sin(x),-0.75,0.75);
    
    # Cost function
    # C(m,b) = sqrt(sum((m*x.+b - y).^2))

    C(m,b) = norm(m*x-y+b*ones(size(x)))

    # Partial derivatives of the cost function

    # Cm(m,b) = sum(2.0.*x.*(m*x.+b - y))
    # Cb(m,b) = sum(2.0.*(m*x.+b - y))

    Cm(m,b) = sum(2.0*x'*(m*x-y+b*ones(size(x))))
    Cb(m,b) = sum(2.0*(m*x-y+b*ones(size(x))))
    
    scatter_plot=scatter(x, y)
    
    step_size = 0.0001

    m,b = [-5, 10.0]

    flag = true

    while flag
        C_current = C(m,b)
#         println("m: $m, b: $b, C(m,b): $(C(m,b))")
        dm,db = -step_size*[Cm(m,b), Cb(m,b)]
        m,b = m+dm, b+db
        
        IJulia.clear_output(true)
        line_plot = plot(x, x->m*x+b)
        scatter_plot = scatter!(x, y)

        Plots.display(scatter_plot)

        sleep(0.0025)
        C_new = C(m,b)
        if abs(C_current - C_new) < 1E-4
            flag = false
        end
    end
end
```
