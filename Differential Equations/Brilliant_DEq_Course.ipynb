```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dxdt(x,t):
    return t*x**2

t = np.linspace(0,4,100)
```

```python
tq = np.arange(0, 2, 0.2)
xq = np.arange(-2, 2, 0.2)

T, X = np.meshgrid(tq, xq)
U = T
V = 0.4*T*X**2

U = U / np.sqrt(U**2 + V**2)
V = V / np.sqrt(U**2 + V**2)
```

```python
t = np.linspace(0,2,100)

x1 = odeint(dxdt, 0.5, t)
x2 = odeint(dxdt, -2, t)
```

```python
fig, ax = plt.subplots(figsize=(9,6))
ax.set_xlim(0,2)
ax.set_ylim(-2,2)

q = ax.quiver(T,X,U,V, width=0.002, units='width')
ax.plot(t,x2)
ax.plot(t,x1)
```
