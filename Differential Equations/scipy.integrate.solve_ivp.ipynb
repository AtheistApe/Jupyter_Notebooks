<p align="center"><h1>Using the "scipy.integrate.solve_ivp()" function to solve differential equations</h1></p>

[Documentation](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html#scipy.integrate.solve_ivp)

The function is called as:
```
scipy.integrate.solve_ivp(fun, t_span, y0, ...)
```

````python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def exp_decay(t,y):
    return -0.5*y
````

````python
sol1 = solve_ivp(exp_decay, [0,10], [2,4,8])
````

````python
print(sol1.t) # time span values
````

````python
print(sol1.y)
````

````python
for row in sol1.y[:,]:
    plt.scatter(sol1.t, row)
````

````python
Y = np.linspace(0,10,10)
sol2 = solve_ivp(exp_decay, [0,10], Y, max_step=0.1, dense_output=True)
````

````python
for row in sol2.y[:,]:
    plt.plot(sol2.t, row)
````

````python
def upward_cannon(t, y):
    return [y[1], -0.5]

def hit_ground(t, y):
    return y[0]

hit_ground.terminal = True
hit_ground.direction = -1

sol = solve_ivp(upward_cannon, [0,100], [0,10],
               events = hit_ground)

print(sol.t_events)
print(sol.t)
````
