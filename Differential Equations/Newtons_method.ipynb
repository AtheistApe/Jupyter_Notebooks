```python
# T'(t) = F(T(t))

# Differential equation for the changing temperature of an object over time.

def F(T0, T_air, k=-1):
    h = 0.01
    t = 0
    T = T0
    ti = [t]
    Ti = [T]
    
    for i in np.arange(0,10,h):
        t += h
        T += h*k*(T - T_air)
        
        ti.append(t)
        Ti.append(T)
        
    return ti, Ti
```

```python
plt.plot(*F(100,20))
plt.plot(*F(100,30))
plt.plot(*F(100,150))
plt.plot(*F(100,100))
```

```python
# T'(t) = F(T(t))

# Differential equation for population growth limited by resources.

def F(P0, K, k=.002):
    h = 0.01
    t = 0
    P = P0
    ti = [t]
    Pi = [P]
    
    for i in np.arange(0,8,h):
        t += h
        P += h*k*P*(K-P)
        
        ti.append(t)
        Pi.append(P)
        
    return ti, Pi
```

```python
plt.plot(*F(50, 500))
plt.plot(*F(800, 500))
plt.plot(*F(500,500))
```

```python
fig = plt.figure(figsize=(8,6))

for pop in np.arange(50, 1000, 50):
    plt.plot(*F(pop,500))
```

```python
# T'(t) = F(T(t))

def F(x0):
    h = 0.01
    t = 0
    x = x0
    ti = [t]
    xi = [x]
    
    for i in np.arange(0,6,h):
        t += h
        x += h*np.sin(x)
        
        ti.append(t)
        xi.append(x)
        
    return ti, xi
```

```python
fig = plt.figure(figsize=(12,9))

for i in np.arange(-2*np.pi, 2*np.pi, np.pi/6):
    plt.plot(*F(i))
```
