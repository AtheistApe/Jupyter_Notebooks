```python
def deriv(y,t):
    a = -2.0
    b = -0.1
    return np.array([ y[1], a*y[0]+b*y[1] ])
```

```python
# time intervals
dt = 0.25
tt = np.arange(0, 10, dt)
# initial condition
xx = [0.1]

def f(x):
    return x * (1.-x)

# loop over time
for t in tt[1:]:
    xx.append(xx[-1] + dt * f(xx[-1]))

# plotting
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(tt, xx, '.-')
ta = np.arange(0, 10, 0.01)
ax.plot(ta, xx[0] * np.exp(ta)/(1+xx[0]*(np.exp(ta)-1.)))
plt.xlabel('t')
plt.ylabel('x')
plt.legend(['approximation', 'analytical solution'], loc='best')
```

```python
def f(x,y):                # DE is y'=f(x,y) = x-y for 
   return x-y              # this function f

def yexact(x,x0,y0):       # exact solution y(x) that satisfies y(x0)=y0
   y=y0*np.exp(-x + x0) + (x*np.exp(x) - x0*np.exp(x0) - np.exp(x) + np.exp(x0))*np.exp(-x)
   return y
```

```python
def deriv(x, t):
    return np.array([0.2*x[0]*(1-x[0]) - 0.3*x[0]**2/(1+x[0]**2)])

def slope(t, x):
    return 0.2*x*(1.0-x) - 0.3*x**2/(1.0+x**2)

tmesh=np.linspace(0.,30.,16); xmesh=np.linspace(0.1,1.,19) # x and y mesh values

time = np.linspace(0.0, 30.0, 1000)

fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(1,1,1)

aspect_ratio = 30

for tp in tmesh:
    for xp in xmesh:
        m=slope(tp,xp)
        h=0.5/np.sqrt(1. + (aspect_ratio*m)**2)
        ax.plot([tp-h,tp+h],[xp-m*h,xp+m*h],'g', linewidth=0.5)
        

for xinit in [0.1,0.3,0.5,0.7,0.9,1]:
    x = odeint(deriv, xinit, time)
    plot(time, x[:,0], linewidth=3)

ax.set_aspect(aspect_ratio)
```

```python
def slope_field(slope_fun, axes, xnum = 10, ynum = 10, length = 0.5, linewidth=1):
    lw = linewidth
    aspect = axes.get_aspect()
    xlim = axes.get_xlim()
    ylim = axes.get_ylim()
    xmesh = np.linspace(xlim[0], xlim[1], xnum+1)
    ymesh = np.linspace(ylim[0], ylim[1], ynum+1)
    
    for xp in xmesh:
        for yp in ymesh:
            m = slope_fun(xp, yp)
            h = length/np.sqrt(1. + (aspect*m)**2)
            axes.plot([xp-h,xp+h],[yp-m*h,yp+m*h],'g',lw)
```
