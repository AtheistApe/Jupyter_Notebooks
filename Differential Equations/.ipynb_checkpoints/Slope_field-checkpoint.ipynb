```python
def slope_field(ax, slope_fun, nx_points=10, ny_points=10, length=5):
    """
    Create a slope field

    Parameters
    ----------
    ax : Axes
        The axes to draw to
        
    slope_fun: A function of (x,y) which returns the slope at position (x,y)
    
    nx_points: The number of grid points in the x direction
    
    ny_points: The number of grid points in the y direction
    
    length: The length of the slope lines

    Returns
    -------
    out : list
        list of artists added
    """
    
    xb = ax.get_xbound()
    yb = ax.get_ybound()
    
    x = np.linspace(xb[0],xb[1],nx_points)
    y = np.linspace(yb[0],yb[1],ny_points)
    
    aspect_ratio = 0.2
    
    for xp in x:
        for yp in y:
            m = slope_fun(xp, yp)
            h=0.05/np.sqrt(1. + (aspect_ratio*m)**2)
            ax.plot([xp-h,xp+h],[yp-m*h,yp+m*h],'g', linewidth=1)
            
```

```python
def get_axes_ratio(ax=None):
    """Calculate the aspect ratio of an axes boundary/frame"""
    if ax is None:
        ax = plt.gca()
    fig = ax.figure

    ll, ur = ax.get_position() * fig.get_size_inches()
    width, height = ur - ll

    return height / width

def get_data_ratio(ax=None):
    """Calculate the data limit ratio of an axes"""
    if ax is None:
        ax = plt.gca()

    xscale, yscale = ax.get_xscale(), ax.get_yscale()
    if xscale != yscale:
        raise ValueError('Cannot get data ratio for semilog plot')
    elif xscale == 'linear':
        return ax.get_data_ratio()
    elif xscale == 'log':
        return ax.get_data_ratio_log()
    else:
        raise ValueError(
            "Unknown scale `{}`; expected `'linear'` or `'log'`".format(xscale))

def get_aspect(ax=None):
    """Calculate the aspect ratio of an axes"""
    if ax is None:
        ax = plt.gca()

    return get_axes_ratio(ax) / get_data_ratio(ax)
```

```python
def slope_field(axes, slope_fun, xnum = 10, ynum = 10, length = 0.5, linewidth=1):
    lw = linewidth
#     axes.set_aspect(1)
    xlim = axes.get_xlim()
    ylim = axes.get_ylim()
    xmesh = np.linspace(xlim[0], xlim[1], xnum)
    ymesh = np.linspace(ylim[0], ylim[1], ynum)
    aspect = get_aspect(ax)
#     print(aspect)

    
    for xp in xmesh:
        for yp in ymesh:
            m = slope_fun(xp, yp)
            dx = 0.5 * length/np.sqrt(1.0 + (aspect*m)**2)
            dy = m * dx
            axes.plot([xp-dx,xp+dx],[yp-dy,yp+dy],'g',lw)
```
