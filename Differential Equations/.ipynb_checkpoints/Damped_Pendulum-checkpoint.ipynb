```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

# Physical constants
g = 9.8     # grav. constant
L = 2.0     # length of pendulum (meters)
μ = 0.3     # coefficient of friction

θ_0 = np.pi/3     # initial position (60 degrees)
θ_dot_0 = 8       # initial angular velocity

# Definition of ODE
def get_theta_double_dot(θ, θ_dot):
    θ_ddot = -μ * θ_dot - (g/L) * np.sin(θ)
    return θ_ddot

# Solution to the differential equation
def theta(t):
    θ = θ_0
    θ_dot = θ_dot_0
    Δt = 0.01     # time step
    for time in np.arange(0, t, Δt):
        θ_double_dot = get_theta_double_dot(θ, θ_dot)
        θ += θ_dot * Δt
        θ_dot += θ_double_dot * Δt
    return (θ, θ_dot)
    
```

```python
t = np.arange(0,40,0.01)
th = [theta(ts)[0] for ts in t]
th_dot = [theta(ts)[1] for ts in t]
```

```python
plt.figure(figsize=(10,6))
plt.plot(th,th_dot)
```

```python
def f(Y, t):
    # y1 = θ, y2 = θ'
    y1, y2 = Y
    g = 9.8
    L = 2.0
    mu = 0.3
    
    return [y2, -1*mu*y2 - (g/L)*np.sin(y1)]

y1 = np.linspace(2.0, 25.0, 46)
y2 = np.linspace(-8.0, 8.0, 32)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
    
    
fig, ax = plt.subplots()

ax.set_aspect(1)
Q = plt.quiver(Y1, Y2, u, v, color='b', scale=150)


# ax.xlabel('$y_1$')
# ax.ylabel('$y_2$')

plt.xlim([2, 25])
plt.ylim([-4, 4])
# plt.savefig('images/phase-portrait.png')
```
