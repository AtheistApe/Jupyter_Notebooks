```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

# Examples

## Example 1
Solve the differential equation
\begin{equation}
\tau\frac{dy}{dt} = -y + ku
\end{equation}
where $\tau$ is the time constant, $k$ is the gain and $u$, the input, is a unit step function.

```python
def model(y, t, k, u):
    tau = 5.0
    dydt = (-y + k*u)/tau
    return dydt

t = np.linspace(0,10,100)
k = 2.0
u = 1.0
```

```python
y = odeint(model, 0, t, args=(k,u))
```

```python
plt.plot(t, y, 'r-', label='solution')
plt.legend()
```

## Example 2
We solve the same DEq as in example 1 but want to change the value of $u$ at different time steps, i.e. $u = u(t)$.

```python
def model(y, t, k, u):
    tau = 5.0
    dydt = (-y + k*u)/tau
    return dydt

t = np.linspace(0,10,100)
k = 2.0

# Here we make the modifications to example 1

u = np.zeros(len(t)) # u is an array of zeros matching the number of elements is ts

u[30:70] = 1.0 # make u into a step function.

y0 = 0 # Initial value y(0)

# We want these u values to sync with the corresponding time values

ys = np.zeros(len(t))
ys[0] = y0

for i in range(len(t)-1):
    ts = [t[i], t[i+1]] # 
    y = odeint(model, y0, ts, args=(k,u[i]))
    y0 = y[1]
    ys[i+1] = y0
```

```python
plt.plot(t, ys, 'b-', label='step function input')
plt.legend()
```

Here's another way of handling a step input function $u(t)$.

```python
def model(y, t, k):
    def u(t):
        if t < 3 or t > 7:
            return 0
        else:
            return 1.0
    
    tau = 5.0
    dydt = (-y + k*u(t))/tau
    return dydt

t = np.linspace(0,10,100)

k = 2.0
```

```python
y = odeint(model, 0, t, args=(k,))
```

```python
plt.plot(t, ys, 'r-', label='original method')
plt.plot(t, y, 'b--', label='alternate method')
plt.legend()
```
