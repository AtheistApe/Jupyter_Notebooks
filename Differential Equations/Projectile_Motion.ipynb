```julia
using DifferentialEquations
using Plots
# pyplot()
# inspectdr()
# gr()
# pgfplots()
```

```julia
function projectile1(du,u,p,t)
    # u[1] = vx, u[2] = vy, u[3] = x, u[4] = y
    
    du[1] = -c*u[1]
    du[2] = -c*u[2]-g
    
    du[3] = u[1]
    du[4] = u[2]
end

g = 9.8
c = 0.25
# c = 0

u0 = [100, 50, 0.0, 0.0]
tspan = (0.0, 8.0)

prob1 = ODEProblem(projectile1, u0, tspan)
sol1 = solve(prob1);
```

```julia
plot(sol1, vars=(3,4))
```

```julia
function projectile2(du,u,p,t)
    # u[1] = vx, u[2] = vy, u[3] = x, u[4] = y
    
    function c(y)
        if y > ymax
            return 0
        else
            return cmax/(m*ymax^2)*(y-ymax)^2
        end
    end
    
    du[1] = -c(u[4])*u[1]
    du[2] = -c(u[4])*u[2]-g
    
    du[3] = u[1]
    du[4] = u[2]
end

g = 9.8
m = 1.0


cmax = 0.25
ymax = 200


u0 = [100, 50, 0.0, 0.0]
tspan = (0.0, 8.0)

prob2 = ODEProblem(projectile2, u0, tspan)
sol2 = solve(prob2);
```

```julia
plot(sol1, vars=(3,4))
plot!(sol2, vars=(3,4))
```

```julia
function projectile(du,u,p,t)
    # u[1] = vx, u[2] = vy, u[3] = x, u[4] = y
    
    du[1] = 0
    du[2] = -gM/(R+u[4])^2
    
    du[3] = u[1]
    du[4] = u[2]
end


G = 6.674e-11 # m^3/(kg s^2)
M = 5.972e24 # kg
R = 6.378e6 # m

gM = G*M
m = 1.0

u0 = [2.17*1.0E4, 2.17*5.0E3, 0.0, 0.0]
tspan = (0.0, 5.0*1.0E4)

prob = ODEProblem(projectile, u0, tspan)
sol = solve(prob,abstol=1e-8,reltol=1e-8);
```

```julia
plot(sol, vars=(3,4))
```

```julia
function periodic(du,u,p,t)
    du[1] = -u[2]
    du[2] = u[1]
end

u0 = [1.0, 0.0]
tspan = (0.0, 6.28)

prob = ODEProblem(periodic, u0, tspan)
sol = solve(prob)

plot(sol, vars=(1,2), aspect_ratio=:equal)
```
