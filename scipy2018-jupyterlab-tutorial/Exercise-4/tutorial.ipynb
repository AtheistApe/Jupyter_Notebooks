```python
import matplotlib.pyplot as plt

img = plt.imread('/home/gary/Projects/SciPy_2019/Numpy-Tutorial-SciPyConf-2019/exercises/filter_image/dc_metro.png')

def smooth(img):
    center = img[1:-1, 1:-1]
    top = img[:-2, 1:-1]
    bottom = img[2:, 1:-1]
    left = img[1:-1, :-2]
    right = img[1:-1, 2:]

    blurred = (top + left + center + right + bottom) / 5
    return blurred

blurred = smooth(img)

difference = img[1:-1,1:-1] - blurred

plt.imshow(img, cmap=plt.cm.hot)
plt.figure()
plt.imshow(blurred, cmap=plt.cm.hot)
plt.show()
plt.figure()
plt.imshow(difference)
```

```python
img = plt.imread('/home/gary/Projects/SciPy_2019/Numpy-Tutorial-SciPyConf-2019/exercises/filter_image/dc_metro.png')

for i in range(50):
    img = smooth(img)
    
plt.imshow(img, cmap=plt.cm.hot)
```
